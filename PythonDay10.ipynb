```python
# 데이터 셋을 훈련용 train 80% 과 검증용 test  20% 으로 분할
```

```python
from sklearn.model_selection import train_test_split

# 문제지 (X : 대문자 X, 관용적인 표현)
X_data = df.loc[:, 'petal_length':'petal_width']
# 정답지
y_data = df.loc[:, 'target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size = 0.2,
                                                    shuffle = True,
                                                    random_state=10)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# KNN 알고리즘
# K-근접 이웃 또는 KNN(K-Nearest Neighbor) 이라고 불리우는 머신러닝 알고리즘

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 3)
knn.fit(X_train, y_train)

# 검증
y_knn_pred = knn.predict(X_test)

# 평가
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_knn_pred)

# n_neighbors= 3     0.9666666666666667
# n_neighbors= 4     1.0
# n_neighbors= 6     1.0
```

```python
# SVM 알고리즘

# 모든 데이터를 공간 내의 좌표에 점으로 표시하고 각 데이터가 속하는 클래스별로 군집을 이룬다라고 가정한다.
# 각 군집 사이의 거리를 최대한 머리 유지하는 선을 찾는다.
# 각 군집을 서로 확연하게 구분할수 있다.

# 학습 -> 모델 생성
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf') # Radial Basis Function 항상 rbf로 사용
svc.fit(X_train, y_train) # 훈련용 데이터 120개 로 학습

# 검증
y_svc_pred = svc.predict(X_test)

# 평가
accuracy_score(y_test, y_svc_pred)
```

```python
# 의사 결정 나무 알고리즘

# 학습 -> 모델 생성
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth = 3, random_state=10)
dtc.fit(X_train, y_train)
# 검증

y_dtc_pred = dtc.predict(X_test)

# 평가
accuracy_score(y_test, y_dtc_pred)

# max_depth = 3   0.9333333333333333
# max_depth = 5   0.9
# max_depth = 10   0.9666666666666667
# max_depth = 20   0.9666666666666667
# max_depth = 30   0.9666666666666667
```
